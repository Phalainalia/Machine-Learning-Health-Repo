## Importaciones

```python
# first neural network with keras tutorial
from numpy import loadtxt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import backend as K
import tensorflow as tf
import sklearn.metrics
```

```python
# load the dataset
# Drive File: dataset = loadtxt('/content/drive/MyDrive/Urestia/indiabetes.csv', delimiter=',')
dataset = loadtxt('/Data/Diabetes.csv', delimiter=',')
# split into input (X) and output (y) variables
X = dataset[:,0:8]
y = dataset[:,8]
```

```python
def keras_f1_score(y_true, y_pred):
    # Cast to float32 for compatibility
    y_true = K.cast(y_true, 'float32')
    # Round predictions to get binary output (0 or 1)
    y_pred = K.cast(K.round(y_pred), 'float32')

    # Calculate True Positives, False Positives, and False Negatives using Keras backend operations
    # TP: y_true is 1 and y_pred is 1
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    # PP: y_true is 1
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    # PR: y_pred is 1
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))

    # Calculate Precision and Recall
    # Use K.epsilon() to avoid division by zero
    precision = true_positives / (predicted_positives + K.epsilon())
    recall = true_positives / (possible_positives + K.epsilon())

    # Calculate F1 Score
    # Use K.epsilon() to avoid division by zero
    f1_score = 2 * ((precision * recall) / (precision + recall + K.epsilon()))

    return f1_score
```

## Modelo Original

### CreaciГіn de Clase Secuencial

```python
# define the keras model
model = Sequential()
model.add(Dense(12, input_shape=(8,), activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

### CompilaciГіn del Modelo

```python
# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

### ActivaciГіn del Modelo

```python
# fit the keras model on the dataset
model.fit(X, y, epochs=150, batch_size=10)
```

### EvaluaciГіn del Modelo

```python
# evaluate the keras model
_, accuracy = model.evaluate(X, y) #Discards the loss with a garbage "_" when generated
print('Accuracy: %.2f' % (accuracy*100))
```

### PredicciГіn del Modelo

```python
# make class predictions with the model
predictions = (model.predict(X) > 0.5).astype(int)
# summarize the first 5 cases
for i in range(5):
	print('%s => %d (expected %d)' % (X[i].tolist(), predictions[i], y[i]))
```

```python
f1 = sklearn.metrics.f1_score(y.astype(int), predictions.flatten())
print('Model F1 Score: %.2f' % (f1*100), '%')
```

## Modelo Modificado

### Nueva Capa Neuronal

```python
# define the keras model
model = Sequential()
model.add(Dense(12, input_shape=(8,), activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(4, activation='relu')) # Se agrega nueva Capa de 4 neuronas al modelo
model.add(Dense(1, activation='sigmoid'))
```

### CompilaciГіn del Modelo

```python
# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# fit the keras model on the dataset
model.fit(X, y, epochs=120, batch_size=12) # Se cambia el learning a 180 ciclos del dset y Delta_peso cada 16 datapoints
```

### EvaluaciГіn y PredicciГіn del Modelo

```python
# evaluate the keras model
_, accuracy = model.evaluate(X, y, verbose=0) # Evaluate again to get the Keras F1 score if needed, or simply avoid overwriting in the previous cell
print('Accuracy: %.2f' % (accuracy*100))
```

```python
predictions = (model.predict(X) > 0.5).astype(int)
f1 = sklearn.metrics.f1_score(y.astype(int), predictions.flatten())
print('Model F1 Score: %.2f' % (f1*100), '%')
```
